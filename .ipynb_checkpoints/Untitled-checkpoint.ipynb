```python
import yfinance as yf
import pandas as pd
import numpy as np
import ta
print("yfinance:", yf.__version__)
print("pandas:", pd.__version__)
```

```python
df = yf.download("AAPL", start="2015-01-01", end="2025-01-1")
df.head()
```

```python
# Flatten columns ให้เหลือชั้นเดียว
df.columns = df.columns.get_level_values(0)

# เหลือแค่ Close ก่อน (ใช้สร้าง features)
df = df[["Open", "High", "Low", "Close", "Volume"]]

# ตรวจสอบ
print(df.shape)
print(df.dtypes)
df.head()
```

```python
from ta.momentum import RSIIndicator
print("ta พร้อมใช้งาน ✓")
```

```python
close = df["Close"]
high = df["High"]
low = df["Low"]

# RSI — momentum (0-100, >70=overbought, <30=oversold)
df["rsi_14"] = ta.momentum.RSIIndicator(close, window=14).rsi()

# MACD — trend direction
macd = ta.trend.MACD(close)
df["macd"]      = macd.macd()
df["macd_signal"] = macd.macd_signal()

# Bollinger Bands — volatility range
bb = ta.volatility.BollingerBands(close, window=20)
df["bb_upper"] = bb.bollinger_hband()
df["bb_lower"] = bb.bollinger_lband()
df["bb_width"] = bb.bollinger_wband()

# Moving Averages — trend
df["sma_20"]  = ta.trend.SMAIndicator(close, 20).sma_indicator()
df["sma_200"] = ta.trend.SMAIndicator(close, 200).sma_indicator()

df.tail()
```

```python
#  สร้าง Normalized Features

#ทำไมต้อง normalize? เพราะ `sma_200 = 210` กับ `rsi = 57` มี scale ต่างกันมาก — model จะงงถ้าใช้ raw values

bb_range = (df["bb_upper"] - df["bb_lower"]).replace(0, float("nan"))
df["bb_position"] = (df["Close"] - df["bb_lower"]) / bb_range

# Close สูงกว่า SMA กี่ %?
df["close_over_sma20"]  = df["Close"] / df["sma_20"]
df["close_over_sma200"] = df["Close"] / df["sma_200"]

# RSI normalize เป็น 0-1
df["rsi_norm"] = df["rsi_14"] / 100

df[["bb_position", "close_over_sma20", "close_over_sma200", "rsi_norm"]].tail()    
```

```python
HORIZON = 5  # ทำนาย 5 วันข้างหน้า

# ราคาปิดอีก 5 วันข้างหน้า
future_close = df["Close"].shift(-HORIZON)

# Target: 1 = ขึ้น, 0 = ลง
df["target"] = (future_close > df["Close"]).astype(int)

# ดูสัดส่วน UP vs DOWN
print(df["target"].value_counts())
print(f"\nUP  : {df['target'].mean():.1%}")
print(f"DOWN: {1 - df['target'].mean():.1%}")
```

```python
FEATURE_COLS = [
    "rsi_norm",
    "macd", "macd_signal",
    "bb_position", "bb_width",
    "close_over_sma20", "close_over_sma200",
]

# ตัด NaN ออก (ช่วงแรกๆ indicators ยังคำนวณไม่ได้)
df_clean = df[FEATURE_COLS + ["target"]].dropna()

print(f"ข้อมูลหลัง dropna: {df_clean.shape}")
print(f"NaN เหลือ: {df_clean.isna().sum().sum()}")
```

```python
# ตัดตาม "เวลา" — ไม่ใช่ random!
# ถ้า random จะเกิด data leakage (ใช้อนาคตทำนายอดีต)
SPLIT_YEAR = 2023

train = df_clean[df_clean.index.year < SPLIT_YEAR]
test  = df_clean[df_clean.index.year >= SPLIT_YEAR]

X_train = train[FEATURE_COLS]
y_train = train["target"]
X_test  = test[FEATURE_COLS]
y_test  = test["target"]

print(f"Train: {X_train.shape} | {train.index.min().date()} → {train.index.max().date()}")
print(f"Test : {X_test.shape}  | {test.index.min().date()} → {test.index.max().date()}")
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=300,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    random_state=42,
    eval_metric="logloss",
)

model.fit(X_train, y_train)
print("Train เสร็จ ✓")
```

```python
from sklearn.metrics import (
    accuracy_score,
    average_precision_score,
    roc_auc_score,
    classification_report,
)

# Predict
y_proba = model.predict_proba(X_test)[:, 1]  # ความน่าจะเป็น UP
y_pred  = (y_proba >= 0.5).astype(int)

# Metrics
print(f"Accuracy : {accuracy_score(y_test, y_pred):.3f}")
print(f"AUC-ROC  : {roc_auc_score(y_test, y_proba):.3f}")
print(f"PR-AUC   : {average_precision_score(y_test, y_proba):.3f}")
print()
print(classification_report(y_test, y_pred))
```
