```python
import yfinance as yf
import pandas as pd
import os

os.makedirs("../../data", exist_ok=True)  # สร้างโฟลเดอร์ data
```

```python
SYMBOLS = [
    "AAPL", "MSFT", "NVDA",   # Tech
    "JPM", "BAC",              # Finance  
    "TSLA", "AMZN",            # Growth
]

START = "2015-01-01"
END   = "2025-01-01"
```

```python
for symbol in SYMBOLS:
    df = yf.download(symbol, start=START, end=END, progress=False)
    
    # Flatten columns
    df.columns = df.columns.get_level_values(0)
    df = df[["Open", "High", "Low", "Close", "Volume"]]
    
    # บันทึกเป็น CSV
    df.to_csv(f"../data/raw/{symbol}.csv")
    print(f"✓ {symbol}: {len(df)} วัน")
```
